```python
data = np.recfromcsv(filename,
                     usecols=[4,6,7,8,9,11,14,15],
                     comments='>',
                     names=['flight','startdate', 'starttime','enddate','endtime',
                            'alt','lat','lon'])
print len(data)
print data[0]
```

```python
home = (39.33149, -76.62356)
setup(figsize=(6,6),
      ylabel='Lat', 
      xlabel='Long')
sc = pylab.scatter(data['lon'], data['lat'], 
                   c=data['alt'], marker='.', edgecolor='none',
                   vmin=0, label='Altitude')
pylab.plot(*home[::-1], marker='s', color='w', markersize=7, 
           lw=4, markeredgecolor='k')
_ = icolorbar(sc)
```

```python
location = LatLon.LatLon(*home)
distance = np.array([location.distance(LatLon.LatLon(la,lon))
                     for la,lon in zip(data['lat'], data['lon'])])
heading = np.array([location.heading_initial(LatLon.LatLon(la,lon))
                     for la,lon in zip(data['lat'], data['lon'])])
altitude = data['alt']*0.0003048
# distance = np.array( LatLon()
# # home = 39.331003, -76.616077
# home = 
# home_utm = np.array(utm.from_latlon(*home)[:2])

# def from_latlon(lat,lon):
#     '''returns x,y in m'''
#     lat,lon = map(np.atleast_1d, (lat,lon))
#     out = [(np.array(utm.from_latlon(la,lo)[:2])-home_utm)/1e3
#            for la,lo in zip(lat,lon)]
#     return map(np.array, zip(*out))
```

```python
theta = -np.deg2rad(heading)+np.pi/2.
x = distance*np.cos(theta)
y = distance*np.sin(theta)
```

```python
dates = date2num([parser.parse(data['startdate'][i]+' '+data['starttime'][i])
                  for i in range(len(data))])
ndays = (dates - np.min(dates))
ndates = ndays/(np.max(ndays)-np.min(ndays))

hours = (dates%1)*24.0
dayint = np.array(map(int, ndays))
```

```python
pylab.figure(figsize=(12,12))
ax = pylab.subplot(polar=True)
ax.set_theta_direction(-1)
ax.set_theta_zero_location('N')
# ax.set_theta_offset(np.pi/2) 
# ax.set_yl
# rlabel_position(np.pi/4)
ax.set_xlabel('heading')
ax.set_ylabel('direction [km]')
# setup(ax=ax, xlabel='heading', ylabel='distance [km]')
_ = ax.plot(np.deg2rad(heading), distance, '.', alpha=0.1)
```

```python
from mpl_toolkits.axes_grid.axislines import SubplotZero
from matplotlib.ticker import MultipleLocator, FuncFormatter

def setup_ax():
    fig = pylab.figure(figsize=(12,12))
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)

    for d in ax.axis:
        ax.axis[d].set_visible(d.endswith("zero"))
    ax.set_aspect('equal')
    tick_format = lambda x, i: '' if x == 0.0 else '%.0f' % x
    for a in [ax.xaxis, ax.yaxis]:
        a.set_minor_locator(MultipleLocator(10))
        a.set_major_formatter(FuncFormatter(tick_format))

    
    for r in range(0,200,25):
        lab = '{:0.0f} km'.format(r)
        if r == 0:
            lab='home'
    #     text(-r*np.cos(45),-r*np.sin(45),lab, clip_on=True,
    #          va='center', ha='center', outline=True)
        ax.add_artist(pylab.Circle((0,0), r, alpha=1-r/200.0,
                                   facecolor='none', edgecolor='0.4'))
```

```python
setup_ax()
den = density(x,y, 
              extent=[-200,50,-200,150],
              label='Log num aircraft', nbins=300, 
              logvrange=True, colorbar=False, cmap=pylab.cm.Greys)
icolorbar(den, loc=2)
setup(aspect='equal', grid=False)
pylab.tight_layout()
```

```python
setup_ax()
den = density(x,y,altitude, 
              extent=[-200,50,-200,150],
              label='average altitude [km]', nbins=900, 
              maskzeros=True, colorbar=False, cmap=pylab.cm.jet)
icolorbar(den, loc=2)
setup(aspect='equal', grid=False)
pylab.tight_layout()
```

```python
setup_ax()
den = density(x,y,hours, 
              extent=[-200,50,-200,150],
              label='hour of the day', nbins=900, 
              maskzeros=True, colorbar=False, cmap=pylab.cm.jet)
icolorbar(den, loc=2)
setup(aspect='equal', grid=False)
pylab.tight_layout()
```

```python
# approximate normalized to window

theta2 = -np.deg2rad(heading)+np.pi/2. - np.pi/2
# theta2 = -np.deg2rad(heading)-np.pi/4
window_distance = distance*np.sin(theta2)/2
xprime = distance*np.cos(theta2) / window_distance
# xprime = np.rad2deg(theta2)
yprime = altitude / window_distance

ii = np.where( (window_distance > 0) &
               (altitude > 0))
setup(figsize=(10,10),
      xlabel='angle relative to center', xr=[-5,5], 
      ylabel='height above horizon', yr=[0,3])
sc = pylab.scatter(xprime[ii], yprime[ii], c=altitude[ii], s=3,
                   lw=0, edgecolor='none', cmap=pylab.cm.jet)
icolorbar(sc)
pylab.tight_layout()
```

```python
setup(figsize=(12,6), subplt=(1,2,1),
      xlabel='distance [km]', ylabel='number')
_ = hist(distance, 100, filled=True, lw=0, alpha=0.7)
setup(subplt=(1,2,2),
      xlabel='altitude [km]', ylabel='number')
_ = hist(altitude, 100, filled=True, lw=0, alpha=0.7)
pylab.tight_layout()
```

```python
setup(figsize=(12,6), xr=[0,24], xtickv=np.arange(0,25,4),
      xlabel='hour of the Day',
      ylabel='number of aircraft per hour')

l,v = None,None
for d in sorted(np.unique(dayint)):
    ii = np.where(d == dayint)[0]
    l,v = hist(hours[ii], np.arange(0,25,1), 
               color=pylab.cm.spectral(d*1.0/(0.3+np.max(dayint))),
               filled=True, lw=0, alpha=0.5, 
               bottom=v, label='day: {}'.format(d+1))
legend(loc=2)
```
